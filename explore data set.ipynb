# explore drug drug similarity network

```python
file_path = './NE Dataset/train_ddi.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

```python
# I will find how many directed edges are here.
check_dict = {}
for i in range(967801):
    first = df['Drug'][i]
    second = df['Target'][i]
    key = (first, second)
    if key in check_dict.keys():
        check_dict[key] += 1
    else:
        check_dict[key] = 1
        
directedCount = 0
for i in range(967801):
    first = df['Drug'][i]
    Second = df['Target'][i]
    revKey = (second, first)
    if revKey in check_dict.keys():
        continue
    else:
        directedCount += 1

print(f"Number of directed edges present in the Drug Drug Similarity Network is: {directedCount}")
```
